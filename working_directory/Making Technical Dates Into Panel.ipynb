# What specifications am I creating?

### 1. 2x2

This will compare the 1990 outcomes of NPPs under construction in 1980. Ones permanently suspended (i.e. cancelled) versus ones completed prior to or during 1990.

For the 2x2, I need under construction for 1980, and variables for cancellation and if under commercial operation by 1990. The dataset will be created by subsetting first to the tracts within the evacuation radiuses of NPPs under construction in 1980, then limiting to those completed by 1990 or permanently suspended. 

### 2. Two-Ways Fixed Effects

This will go from 1970 to 2000, sampling at every decade. NPPs will be either `under_construction`, in `commercial_operation`, or `retired`. 

For TWFE, I will have variables for under_construction, commercial_operation, and retired, for 1970, 1980, 1990, and 2000. 



So in sum, it will be `under_construction`, `commercial_operation`, and `retired`,  for 1970, 1980, 1990, and 2000, along with `permanently_suspended` for those permanently suspended in the 1980s and not resumed.
